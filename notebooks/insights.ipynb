```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

# style
sns.set_style("whitegrid")

# path
base_path = Path().resolve().parent
```

```python
df = pd.read_csv(base_path / "ukraine-book-publishing-analysis/data/processed/ukraine_books.csv")

df.head() 
```

```python
plt.figure()

plt.stackplot(
    df['year'],
    df['ukr_copies_mln'],
    df['rus_copies_mln'],
    df['other_copies_mln'],
    labels=['Ukrainian', 'Russian', 'Other']
)

plt.legend(loc='upper left')
plt.title('Language distribution of book publishing')
plt.xlabel('Year')
plt.ylabel('Million copies')

plt.show()
```

```python
df['rus_share_pct'] = (df['rus_copies_mln'] / df['total_copies_mln'] * 100).round(2)
df['other_share_pct'] = (df['other_copies_mln'] / df['total_copies_mln'] * 100).round(2)
```

```python
plt.figure()

plt.plot(df['year'], df['ukr_share_pct'], label='Ukrainian')
plt.plot(df['year'], df['rus_share_pct'], label='Russian')

plt.legend()
plt.title('Language share (%)')

plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df['year'], df['total_titles'])
ax1.set_ylabel('Titles')

ax2 = ax1.twinx()
ax2.plot(df["year"], df["avg_print_run"])
ax2.set_ylabel("Average print run")

plt.title("Titles vs Average print run")
plt.show()
```

```python
plt.figure()

plt.plot(df['year'], df['fiction_titles'], label='Fiction')
plt.plot(df['year'], df['non_fiction_titles'], label='Non-fiction')
plt.plot(df['year'], df['edu_titles'], label='Education')

plt.legend()
plt.title('Titles by category')

plt.show()
```

## Key Insights

1. The data shows a sharp decline in book publishing output after 1991, following the collapse of the Soviet Union and the transition to a market economy. The lowest levels were observed around 1994. Both total print runs and Russian-language publications decreased significantly during this period.

2. Over time, the structure of the market changed. The share of Ukrainian-language publications increased steadily, while the share of Russian-language books declined. This shift likely reflects a combination of domestic market development, reduced imports, and changing consumer preferences.

3. The Ukrainian book publishing market is highly sensitive to external shocks. Significant declines are visible during major кризис periods, including 2014 (start of war), 2020 (COVID-19 pandemic), and 2022 (full-scale invasion).

4. In recent years, Ukrainian has become the dominant language in book publishing. Other languages, including Russian, represent a relatively small share of total print runs.

5. The average print run per title has decreased significantly compared to the early 1990s, indicating a structural shift from mass printing to a more fragmented market with a larger number of titles and smaller print runs.
